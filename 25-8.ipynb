```python
# part 1

import math
from typing import List, Dict, NamedTuple
from collections import namedtuple

Point3 = NamedTuple('Point3', [('x', int), ('y', int), ('z', int)])
def add(self: Point3, other: Point3) -> Point3:
    return Point3(self.x + other.x, self.y + other.y, self.z + other.z)
def sub(self: Point3, other: Point3) -> Point3:
    return Point3(self.x - other.x, self.y - other.y, self.z - other.z)
def mul(self: Point3, scalar: int) -> Point3:
    return Point3(self.x * scalar, self.y * scalar, self.z * scalar)
def repr(self: Point3):
    return f"<{self.x}, {self.y}, {self.z}>"
Point3.__add__ = add
Point3.__sub__ = sub
Point3.__mul__ = mul
Point3.__repr__ = repr

def distance(pt1: Point3, pt2: Point3):
    sub = pt1 - pt2
    return math.sqrt(sub.x * sub.x + sub.y * sub.y + sub.z * sub.z)

nodes: List[Point3] = []

for line in input.splitlines():
     x, y, z = list(map(int, line.split(",")))
     nodes.append(Point3(x, y, z))

print ("Number of nodes", len(nodes))

Edge = NamedTuple('Edge', [('pt1', Point3), ('pt2', Point3)])

potentialEdgeList = []

numAppended = 0
printInterval = 50000

for i1, pt1 in enumerate(nodes):
    for pt2 in nodes[i1 + 1:]:
        potentialEdgeList.append(Edge(pt1, pt2))
        numAppended += 1
        if (numAppended % printInterval == 0):
            print ("Num appended:", numAppended)

print ("Number of potential edges:", len(potentialEdgeList))

potentialEdgeList = sorted(potentialEdgeList, key=lambda edge: distance(edge.pt1, edge.pt2))

#print (potentialEdgeList)

import networkx as nx
from networkx.algorithms.components import connected_components

G = nx.Graph()
for node in nodes:
    G.add_node(node)
for i in range(numconnections):
    edge = potentialEdgeList[i]
#    print (edge)
    G.add_edge(edge.pt1, edge.pt2, dist = distance(edge.pt1, edge.pt2))

print (G)

componentLengths = [len(c) for c in sorted(nx.connected_components(G), key=len, reverse=True)]

componentLengths[0] * componentLengths[1] * componentLengths[2]
```
