```python
# part 2

from typing import Set, List, Dict, NamedTuple
from collections import namedtuple

Point = NamedTuple('Point', [('x', int), ('y', int)])
def add(self: Point, other: Point) -> Point:
    return Point(self.x + other.x, self.y + other.y)
def sub(self: Point, other: Point) -> Point:
    return Point(self.x - other.x, self.y - other.y)
def mul(self: Point, scalar: int) -> Point:
    return Point(self.x * scalar, self.y * scalar)
def repr(self: Point):
    return f"<{self.x}, {self.y}>"
Point.__add__ = add
Point.__sub__ = sub
Point.__mul__ = mul
Point.__repr__ = repr

Rect = NamedTuple('Rect', [('first', Point), ('second', Point)])

def oob(pt: Point):
    return (pt.x < 0 or pt.y < 0 or pt.x >= width or pt.y >= height)
up = Point(0, -1)
right = Point(1, 0)
down = Point(0, 1)
left = Point(-1, 0)
directions = [up, right, down, left]

def rectArea(corner1: Point, corner2: Point):
    width = abs(corner1.x - corner2.x) + 1
    height = abs(corner1.y - corner2.y) + 1
    return width * height


corners: List[Point] = []
width = 0
height = 0
for line in input.splitlines():
    x, y = list(map(int, line.split(",")))
    corners.append(Point(x, y))
    width = max(x+2, width)
    height = max(y+2, height)

print (corners, width, height)

sides: Set[Point] = set()

def getDir(pt1: Point, pt2: Point):
    if (pt1.x == pt2.x):
        if (pt1.y < pt2.y):
            dir = down
        else:
            dir = up
    else:
        if (pt1.x < pt2.x):
            dir = right
        else:
            dir = left
    return dir

#for i, pt1 in enumerate(corners):
#    pt2 = corners[i-1]
#    dir = getDir(pt1, pt2)
#    if (i == 0): firstDir = dir
#    if (i == 1): secondDir = dir
#    if (i == 2): thirdDir = dir
#    pt = pt1
#    sides.add(pt)
#    while (pt != pt2):
#        pt += dir
#        sides.add(pt)

#print (sides)

#def floodFill(pt: Point):
#    if (pt in sides): return
#
#    sides.add(pt)
#
#    for dir in directions:
#        nextPt = Point(pt.x + dir.x, pt.y + dir.y)
#        floodFill(nextPt)

def printGrid():
    grid: List[str] = []
    for y in range(height):
        line = ""
        for x in range(width):
            if Point(x, y) in sides:
                line = line + "X"
            else:
                line = line + "."
        grid.append(line)

    for line in grid:
        print (line)

#printGrid()

#def floodFill(pt: Point):
#    global sides
#    nextPts: Set[Point] = set()
#    nextPts.add(pt)
#
#    while len(nextPts) > 0:
#        sides = sides | nextPts
#
#        prevNextPts = nextPts
#        nextPts = set()
#        for pt in prevNextPts:
#            for dir in directions:
#                nextpt = Point(pt.x + dir.x, pt.y + dir.y)
#                if (oob(nextpt)):
#                    raise ValueError
#                if (nextpt not in sides):
#                    nextPts.add(nextpt)

#print (firstDir, secondDir, thirdDir)
#floodFill(corners[2] + secondDir + thirdDir)
#floodFill(Point(1, 1))
#print("Floodfill complete")

#printGrid()

areas: Dict[int, Rect] = {}
for i, pt1 in enumerate(corners):
    for pt2 in corners[i+1:]:
        areas[rectArea(pt1, pt2)] = Rect(pt1, pt2)

areas = dict(sorted(areas.items(), reverse = True))
#print (areas)

Interval = NamedTuple('Interval', [('start', int), ('end', int)])
def contains(self: Interval, num: int):
     return (num >= self.start and num <= self.end) or (num >= self.end and num <= self.start)
def repr(self: Interval):
    return f"<{self.start}, {self.end}>"
Interval.__contains__ = contains
Interval.__repr__ = repr

def isInsidePolygon(pt: Point, debug: bool):
    dirs = directions
    foundDirs = set()
    for i, pt1 in enumerate(corners):
        pt2 = corners[i-1]
        for dir in dirs:
            if dir == left:
                if pt1.x == pt2.x and pt.x >= pt1.x:
                    if pt.y in Interval(pt1.y, pt2.y):
                        if (debug): print ("Found left wall")
                        foundDirs.add(dir)
                        if (debug): print (dirs)
            elif dir == right:
                if pt1.x == pt2.x and pt.x <= pt1.x:
                    if pt.y in Interval(pt1.y, pt2.y):
                        if (debug): print ("Found right wall")
                        foundDirs.add(dir)
                        if (debug): print (dirs)
            elif dir == up:
                if pt1.y == pt2.y and pt.y >= pt1.y:
                    if pt.x in Interval(pt1.x, pt2.x):
                        if (debug): print ("Found up wall")
                        foundDirs.add(dir)
                        if (debug): print (dirs)
            elif dir == down:
                if pt1.y == pt2.y and pt.y <= pt1.y:
                    if pt.x in Interval(pt1.x, pt2.x):
                        if (debug): print ("Found down wall")
                        foundDirs.add(dir)
                        if (debug): print (dirs)
            if (len(foundDirs) == 4):
                return True
    return False

def spansInnerSpace(pt1: Point, pt2: Point, debug: bool):
    dir = getDir(pt1, pt2)
    if (dir == left or dir == right):
        y = pt1.y
        x1 = min(pt1.x, pt2.x)
        x2 = max(pt1.x, pt2.x)
        for i, corner1 in enumerate(corners):
            corner2 = corners[i-1]
            cornerDir = getDir(corner1, corner2)
            if (cornerDir == up or cornerDir == down):
                cornerX = corner1.x
                cornerY1 = min(corner1.y, corner2.y)
                cornerY2 = max(corner1.y, corner2.y)
                if (y > cornerY1 and y < cornerY2 and cornerX > x1 and cornerX < x2):
                    intersection = Point(cornerX, y)
                    if not isInsidePolygon(intersection + left, debug) or not isInsidePolygon(intersection + right, debug):
                        return False

    if (dir == up or dir == down):
        x = pt1.x
        y1 = min(pt1.y, pt2.y)
        y2 = max(pt1.y, pt2.y)
        for i, corner1 in enumerate(corners):
            corner2 = corners[i-1]
            cornerDir = getDir(corner1, corner2)
            if (cornerDir == left or cornerDir == right):
                cornerY = corner1.y
                cornerX1 = min(corner1.x, corner2.x)
                cornerX2 = max(corner1.x, corner2.x)
                if (x > cornerX1 and x < cornerX2 and cornerY > y1 and cornerY < y2):
                    intersection = Point(x, cornerY)
                    if not isInsidePolygon(intersection + up, debug) or not isInsidePolygon(intersection + down, debug):
                        return False

    return True


for area, rect in areas.items():
    pt1 = rect[0]
    pt2 = rect[1]
    pt3 = Point(pt1.x, pt2.y)
    pt4 = Point(pt2.x, pt1.y)
    fullrect = [pt1, pt3, pt2, pt4]
    
    debug = False
    if (area == 24): debug = True
    if debug: print (f"Testing {area}, {rect}")

    if not isInsidePolygon(pt3, debug) or not isInsidePolygon(pt4, debug):
        continue

    valid = True
    for i, ptA in enumerate(fullrect):
        ptB = fullrect[i-1]
        if not spansInnerSpace(ptA, ptB, debug):
            valid = False
            break
    if valid == False: continue

    print(area)
    break

#for key, value in areas:
#    print(f"{key}: {areas[key]}")
```
