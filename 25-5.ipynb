```python
input = """3-5
10-14
16-20
12-18

1
5
8
11
17
32
"""
```

```python
# part 1

from typing import List, Dict, NamedTuple
from collections import namedtuple

Interval = NamedTuple('Interval', [('start', int), ('end', int)])
def contains(self: Interval, num: int):
     return num >= self.start and num <= self.end
def repr(self: Interval):
    return f"<{self.start}, {self.end}>"
Interval.__contains__ = contains
Interval.__repr__ = repr

freshRanges: List[Interval] = []
ids: List[int] = []

parsingRanges = True
for line in input.splitlines():
    if len(line) == 0:
        parsingRanges = False
        continue
    if parsingRanges:
        start, end = map(int, line.split("-"))
        freshRanges.append(Interval(start, end))
    else:
        ids.append(int(line))

print (freshRanges, ids)

sum = 0

for id in ids:
    fresh = False
    for interval in freshRanges:
        if id in interval:
            fresh = True
            break
    
    if fresh == True:
        print(id)
        sum += 1

print (sum)
```

```python
# part 2

from typing import List, Dict, NamedTuple
from collections import namedtuple

Interval = NamedTuple('Interval', [('start', int), ('end', int)])
def contains(self: Interval, num: int):
     return num >= self.start and num <= self.end
def repr(self: Interval):
    return f"<{self.start}, {self.end}>"
Interval.__contains__ = contains
Interval.__repr__ = repr

ranges: List[Interval] = []

for line in input.splitlines():
    if len(line) == 0:
        break
    start, end = map(int, line.split("-"))
    ranges.append(Interval(start, end))


ranges = sorted(ranges, key=lambda r: r.end)
ranges = sorted(ranges, key=lambda r: r.start)

def merge(ranges):
    merged: List[Interval] = []
    last = Interval(0, 0)
    for r in ranges:
        interval = r
        if last.end in r or r.end in last:
            interval = Interval(min(r.start, last.start), max(r.end, last.end))
        else:
            if last.start != 0:
                merged.append(last)
        last = interval
    merged.append(last)

    return merged

merged = ranges

iteration_max = 3
iteration = 0

while iteration < iteration_max:
    merged = merge(merged)

    invalid = False
    for r in merged:
        startAppearCount = 0
        endAppearCount = 0
        copy = merged[:]
        copy.remove(r)
        for o in copy:
            if r.start in o:
                startAppearCount += 1
            if r.end in o:
                endAppearCount += 1
            if startAppearCount > 0:
                print ("Start appears ", startAppearCount, " for ", r, "in", o)
                invalid = True
                break
            if endAppearCount > 0:
                print ("End appears ", endAppearCount, " for ", r, "in", o)
                invalid = True
                break
 
   
    if (invalid == False):
        break

    iteration += 1
    print ("iterating")

for r in merged:
   print(r)

sum = 0
for r in merged:
    sum += r.end - r.start + 1

print(sum)
```
