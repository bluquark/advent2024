```python
input = """123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  
"""
```

```python
# part 1

from typing import List, Dict, NamedTuple

rows: List[List[int]] = []
operators: List[str] = []

for line in input.splitlines():
    row = line.split()
    try:
        rows.append(list(map(int, row)))
    except ValueError:
        operators = row
    
#print(rows)
#print(operators)

sum = 0

#123: row[0][0]
#45: row[1][0]
#6: row[2][0]

#328: row[0][1]
#64: row[1][1]
#98: row[2][1]

for i, o in enumerate(operators):
    result = rows[0][i]
    #print (result)
    for row in rows[1:]:
        num = row[i]
        #print (num)
        if o == "*":
            result *= num
        else:
            result += num
    #print(result)
    sum += result

print (sum)
```
