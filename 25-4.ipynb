```python
# part 1

from typing import List, Dict, NamedTuple
from collections import namedtuple

Point = NamedTuple('Point', [('x', int), ('y', int)])
def add(self: Point, other: Point) -> Point:
    return Point(self.x + other.x, self.y + other.y)
def sub(self: Point, other: Point) -> Point:
    return Point(self.x - other.x, self.y - other.y)
def mul(self: Point, scalar: int) -> Point:
    return Point(self.x * scalar, self.y * scalar)
def repr(self: Point):
    return f"<{self.x}, {self.y}>"
Point.__add__ = add
Point.__sub__ = sub
Point.__mul__ = mul
Point.__repr__ = repr

grid = input.splitlines()

rolls = []
for y, row in enumerate(grid):
    height = y + 1
    for x, column in enumerate(row):
        width = x + 1
        if (column == "@"):
           rolls.append(Point(x, y))

def oob(pt: Point):
    return (pt.x < 0 or pt.y < 0 or pt.x >= width or pt.y >= height)

def hasRoll(pt: Point):
    if (oob(pt)): return False
    return pt in rolls

up = Point(0, -1)
right = Point(1, 0)
down = Point(0, 1)
left = Point(-1, 0)
directions = [up, right, down, left, up + left, up + right, down + left, down + right]
#print (directions)

sum = 0

for r in rolls:
    numAdjacent = 0
    for dir in directions:
       if (hasRoll(r + dir)):
           numAdjacent += 1
    if (numAdjacent < 4):
        sum += 1
    
print (sum)
```
