```python
input = """11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124"""
```

```python
# part 1

sum = 0

for pair in input.split(","):
    first, second = map(int, pair.split("-"))
    #print(first, second)

    for id in range(first, second + 1):
        #print(id)
        s = str(id)
        if (len(s) % 2 == 1): continue
        s1 = s[:len(s) // 2]
        s2 = s[len(s) // 2:]
        if (s1 == s2):
            sum += int(s)
            #print(s)

print (sum)
```

```python
# part 2

invalid = {0}

for pair in input.split(","):
    first, second = map(int, pair.split("-"))
    #print(first, second)

    for id in range(first, second + 1):
        #print(id)
        s = str(id)
        for factor in range(2, len(s) + 1):
            if (len(s) % factor > 0): continue
            match = 0
            interval = len(s) // factor
            s1 = s[:interval]
            remaining = s[interval:]
            while len(remaining) > 0:
                 s2 = remaining[:interval]
                 remaining = remaining[interval:]
                 if (s1 == s2):
                     #print (s, factor, s1, s2)
                     match += 1
            if (match == factor - 1):
                invalid.add(int(s))
                print(s)

sum = 0
for i in invalid:
    sum += i
print (sum)
```
