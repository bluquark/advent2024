```python
# part 2, second attempt

from typing import List, Dict, Tuple, NamedTuple
from collections import namedtuple

longestBank = 1

Bank = NamedTuple('Bank', [('i', int), ('isButton', bool)])

def numBitsSet(self: Bank):
    num = 0
    for sh in range(longestBank):
        if (self.i & (1 << sh) != 0): num += 1

    return num

def joltageSum(joltage: Tuple[int]):
    sum = 0
    for j in joltage:
        sum += j
    return sum

def repr(self: Bank):
    if (self.isButton):
        str = "("
    else:
        str = "["
    for sh in range(longestBank):
        if (self.i & (1 << sh)) == 0:
            str += "."
        else:
            if (self.isButton):
                str += "+"
            else:
                str += "#"
    if (self.isButton):
        str += ")"
    else:
        str += "]"
    return str
Bank.__repr__ = repr

bankList: List[Bank] = []
buttonsList: List[Bank] = []
joltagesList: List[Tuple] = []

for line in input.splitlines():
    bankStr, rest = line.split(' ', maxsplit=1)
    buttonsStr, joltagesStr = rest.split('{') 
    #print (bankStr, buttonsStr, joltagesStr)

    bankStr = bankStr.strip('[]')
    longestBank = max(longestBank, len(bankStr))
    bank = 0
    for lightNo, ch in enumerate(bankStr):
        if ch == '#': bank = bank | (1 << lightNo)
        #print (ch, lightNo, bank)
    #print (Bank(bank))
    bankList.append(Bank(bank, False))

    buttons = []
    for buttonStr in buttonsStr.rstrip().split(' '):
        buttonIndices = tuple(map(int, buttonStr.strip('()').split(',')))
        buttonInt = 0
        for index in buttonIndices:
            buttonInt += 1 << index
        buttons.append(Bank(buttonInt, True))
        #print (toggles)
    buttons.sort(key = numBitsSet, reverse = True)
    buttonsList.append(tuple(buttons))
    
    joltages = tuple(map(int, joltagesStr.strip(' {}').split(',')))
    #print (joltages)
    joltagesList.append(joltages)


#print (bankList, buttonsList, joltagesList)


# https://developers.google.com/optimization/pack/bin_packing
from ortools.linear_solver import pywraplp

def countMinPresses(joltage: Tuple[int], buttons: Tuple[Bank]):
    solver = pywraplp.Solver.CreateSolver('SCIP') 

    vars = []
    for b in buttons:
        vars.append(solver.IntVar(0.0, solver.infinity(), repr(b)))

    print (joltage, buttons)

    for indexJ, jolt in enumerate(joltage):
        varsAffectingJolt = []
        for indexB, button in enumerate(buttons):
            if button.i & (1 << indexJ) != 0:
                varsAffectingJolt.append(vars[indexB])
        #print (jolt, varsAffectingJolt)
        solver.Add(sum(varsAffectingJolt) == jolt)
        
    solver.Minimize(solver.Sum(vars))
    solver.Solve()
    
    return solver.Objective().Value()


totalPresses = 0
for i in range(len(bankList)):
    count = countMinPresses(joltagesList[i], buttonsList[i])
    print (count)
    totalPresses += count

totalPresses


```
