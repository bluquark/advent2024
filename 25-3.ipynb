```python
input = """987654321111111
811111111111119
234234234234278
818181911112111
"""
```

```python
# part 1

sum = 0

for line in input.splitlines():
    bank = []
    for ch in line:
        bank.append(int(ch))
    #print(bank)

    highestIndex = 0
    for index, num in enumerate(bank):
        if num > bank[highestIndex]: highestIndex = index
    
    secondHighest = 0
    if (highestIndex == len(bank) - 1):
        for num in bank[:highestIndex]:
            if num > secondHighest: secondHighest = num
        output = secondHighest * 10 + bank[highestIndex]
    else:
        for num in bank[highestIndex+1:]:
            if num > secondHighest: secondHighest = num
        output = secondHighest + bank[highestIndex] * 10

    print(output)
    sum += output

print(sum)
```

```python
# part 2

sum = 0

for line in input.splitlines():
    bank = []
    for ch in line:
        bank.append(int(ch))
    #print(bank)

    output = []
    remainingDigits = 12
    remainingBank = bank[:]

    while remainingDigits > 0:
        if (remainingDigits > 1):
            selectBank = remainingBank[:-(remainingDigits-1)]
        else:
            # special case because list[:-0] produces an empty list
            selectBank = remainingBank
        highestIndex = 0
        for index, num in enumerate(selectBank):
            if num > remainingBank[highestIndex]: highestIndex = index

        output.append(remainingBank[highestIndex])

        remainingDigits -= 1
        remainingBank = remainingBank[highestIndex+1:]
    
    print(output)

    multiplier = 1
    for num in reversed(output):
        sum += num * multiplier
        multiplier *= 10

print(sum)
```
