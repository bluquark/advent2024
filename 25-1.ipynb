```python
# part 1

arrow = 50
numzeroes = 0

for line in input.splitlines():
    sign = 1
    if (line[0] == "L"):
        sign = -1
    numturns = int(line[1:])

    print(sign, numturns)

    arrow += (sign * numturns)
    print(arrow)
    if (arrow % 100 == 0): numzeroes += 1


print(numzeroes)
```
