```python
input = """.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............
"""
```

```python
# part 1

beamXs = set()

splitCount = 0

for line in input.splitlines():
    for i, ch in enumerate(line):
        if ch == "S":
            beamXs.add(i)
        if ch == "^":
            if (i in beamXs):
                splitCount += 1
                beamXs.remove(i)
                beamXs.add(i - 1)
                beamXs.add(i + 1)

print(splitCount)
```

```python
# part 2

import functools

rows: List[str] = []

for line in input.splitlines():
    rows.append(line)

@functools.lru_cache(maxsize=None)
def countPaths(y:int, x:int):
    if (y == len(rows) - 1): return 1
    if (rows[y][x] == '^'):
        return countPaths(y + 1, x - 1) + countPaths(y + 1, x + 1)
    else:
        return countPaths(y + 1, x)

for i, ch in enumerate(rows[0]):
    if (ch == "S"): startX = i

print (countPaths(1, startX))
```
